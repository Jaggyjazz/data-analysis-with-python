Here we can see that the entire rural row is missing from the data and we can choose to either remove 'West Bengal' or add a row with area as'Rural' and fill other data from a respected source, wherever we need both Rural and Urban areas for comparison of data.

For now, I am dropping the state 'West Bengal' from the data, assuming that this is the only data we are able to work with and additional data on 'West Bengal / Rural' couldn't be found anywhere.

```python
all_areas_df = df[df['State/UT'] != 'West Bengal']
all_areas_df
```

## Trend Analysis of Tobacco Usage "Rural" vs "Urban"

### Separating into a smaller DataFrame

```python
data = all_areas_df[['State/UT','Area','Ever tobacco users (%)','Current tobacco users (%)' ]]
data
```

#### Plotting the data

```python
sns.barplot(x='State/UT', y='Ever tobacco users (%)', hue='Area', data=data)
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(25, 12))

# Second plot (scatter plot + linear regression line)
sns.regplot(x='Ever tobacco users (%)', y='Current tobacco users (%)', data=data, ax=axs[0])
# Second plot (scatter plot + Polynomial regression line)
sns.regplot(x='Ever tobacco users (%)', y='Current tobacco users (%)', data=data, order=2, ax=axs[1])

axs[0].tick_params(axis='x', rotation=90)  # Rotate x-axis labels for readability
axs[1].tick_params(axis='x', rotation=90)  # Rotate x-axis labels for readability

plt.show()
```

```python
fig, ax = plt.subplots()
df.plot(kind='hexbin', x='Ever tobacco users (%)', y='Current tobacco users (%)', gridsize=20, cmap='viridis', ax=ax)
ax.set_facecolor('black')
plt.show()
```
